# Language Model (N-Grams)

#### Vectorizing

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
class NGramLanguageModel: 
    def __init__(self, n):
        self.n = n
        self.vectorizer = CountVectorizer(analyzer='word', ngram_range=(n,n))
        
    def fit_transform(self, data): 
        return self.vectorizer.fit_transform(data)
    
    def transform(self, data): 
        return self.vectorizer.transform(data)
```

```python
document = [
  'She sells sea shells by the sea shore',
  'The shells she sells are surely sea shells',
  'So if she sells sea shells on the sea shore'
]

query = 'Surely she sells sea shells'
```

```python
n = 1   # unigram = 1, bigram = 2, trigram = 3
nGramModel = NGramLanguageModel(n)

document_vector = nGramModel.fit_transform(document)
query_vector = nGramModel.transform([query])
pd.DataFrame(document_vector.toarray(), columns=nGramModel.vectorizer.get_feature_names_out())
```

```python
n = 2     # bigram
nGramModel = NGramLanguageModel(n)

document_vector = nGramModel.fit_transform(document)
query_vector = nGramModel.transform([query])
pd.DataFrame(document_vector.toarray(), columns=nGramModel.vectorizer.get_feature_names_out())
```

```python
n = 3     # trigram
nGramModel = NGramLanguageModel(n)

document_vector = nGramModel.fit_transform(document)
query_vector = nGramModel.transform([query])
pd.DataFrame(document_vector.toarray(), columns=nGramModel.vectorizer.get_feature_names_out())
```

#### Cosine Similarity

```python
def calculate_cosine_similarity(target, query): 
    return cosine_similarity(target, query)
```

```python
similarity = calculate_cosine_similarity(document_vector, query_vector)

data = {
    'document': document, 
    'similarity': similarity.flatten()
}

pd.DataFrame(data) # n=1 (unigram)
```
