```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
class NGramLanguageModel: 
    def __init__(self, n):
        self.n = n
        self.vectorizer = CountVectorizer(analyzer='word', ngram_range=(n,n))
        
    def fit_transform(self, data): 
        return self.vectorizer.fit_transform(data)
    
    def transform(self, data): 
        return self.vectorizer.transform(data)
```

```python
document = [
    'Try to Learning',
    'Natural Language Processing',
    'In Software Laboratory Class'
]

query = 'Try Query the Natural Language Processing Corpus'
```

```python
n = 1    
nGramModel = NGramLanguageModel(n)

document_v = nGramModel.fit_transform(document)
query_v = nGramModel.transform([query])
pd.DataFrame(document_v.toarray(), columns=nGramModel.vectorizer.get_feature_names_out())
```

```python
def calculate_cosine_similarity(target, query): 
    return cosine_similarity(target, query)
```

```python
similarity = calculate_cosine_similarity(document_v, query_v)

data = {
    'document': document, 
    'similarity': similarity.flatten()
}

pd.DataFrame(data) # n=1 (unigram)
```
